```python
%matplotlib inline
import matplotlib.pyplot as plt 
plt.style.use('seaborn-white')
import numpy as np
```

```python
# simple colorbar legend

x = np.linspace(0, 10, 1000) 
I = np.sin(x) * np.cos(x[:, np.newaxis])

plt.imshow(I) 
plt.colorbar();
```

```python
# customize colorbar
#specify colormap using cmap arg

plt.imshow(I, cmap='Blues');
```

```python
#handwritten digits
# load images of handwritten digits 0 through 5
```

```python
from sklearn.datasets import load_digits

digits = load_digits(n_class=6)

fig, ax = plt.subplots(8, 8, figsize=(6, 6))
ax.flat
```

```python
digits = load_digits(n_class=6)

fig, ax = plt.subplots(8, 8, figsize=(6, 6))

for i, axi in enumerate(ax.flat):
    axi.imshow(digits.images[i], cmap='binary')
    axi.set(xticks=[], yticks=[])
```

```python
# Use dimensionality reduction technique such as manifold learning to reduce the dimensionality of the data 
#while maintaining the relationships of interest.
```

```python
# project digits into 2 dimensions using Isomap
from sklearn.manifold import Isomap

iso = Isomap(n_components=2, n_neighbors=15)
projection = iso.fit_transform(digits.data)

# plot results
plt.scatter(projection[:,0], projection[:,1], lw=0.1, c=digits.target, cmap=plt.cm.get_cmap('viridis', 512))
plt.colorbar(ticks=range(6), label='digit value')
plt.clim(-0.5, 5.5)

# there is overlap between 2 and 3 digits indicating it is difficult to distinguish between handwritten 2 and 3.
# Digits like 0 and 1 are easily separated.
```
