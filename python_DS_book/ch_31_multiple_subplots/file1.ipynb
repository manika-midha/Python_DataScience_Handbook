```python
# plt.subplot

for i in range(1, 7):
    '''
      Three integers (*nrows*, *ncols*, *index*). The subplot will take the
      *index* position on a grid with *nrows* rows and *ncols* columns.
      *index* starts at 1 in the upper left corner and increases to the
      right.
    '''
    plt.subplot(2, 3, i)
    '''
    plt.text(x, y, s)
    x,y = The position to place the text
    s = str (the text)
    ha = horizontal alignment
    '''
    plt.text(.5, .5, str((2, 3, i)), fontsize=18, ha='center', va='center');
```

```python
# adjust spacing between plots
fig = plt.figure()
fig.subplots_adjust(hspace=.4, wspace=.4) # the space is 40% of subplot width and height
for i in range(1, 7):
    ax = fig.add_subplot(2, 3, i)
    ax.text(.5, .5, str((2, 3, i)), fontsize=18, ha='center');
```

```python
# example
# create normally distributed data
mean = [0, 0]
cov = [[1, 1], [1, 2]]
rng = np.random.default_rng(1701)
x,y = rng.multivariate_normal(mean, cov, 3000).T

# set up axes with GridSpec
fig = plt.figure(figsize=(6, 6))
grid = plt.GridSpec(4, 4, hspace=.2, wspace=.2)
main_ax = fig.add_subplot(grid[:-1, 1:])
y_hist = fig.add_subplot(grid[:-1, 0], xticklabels=[], sharey=main_ax)
x_hist = fig.add_subplot(grid[-1, 1:], yticklabels=[], sharex=main_ax)

# scatter points on main axis
main_ax.plot(x,y, 'ok', markersize=3, alpha=.2)

# histogram on the attached axes
x_hist.hist(x, 40, histtype='stepfilled', orientation='vertical', color='green')
x_hist.invert_yaxis()

y_hist.hist(y, 40, histtype='stepfilled', orientation='horizontal', color='green')
y_hist.invert_xaxis()
```
