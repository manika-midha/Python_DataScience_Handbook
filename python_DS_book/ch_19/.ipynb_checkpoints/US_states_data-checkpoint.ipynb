```python
pop = pd.read_csv('state-population.csv')
areas = pd.read_csv('state-areas.csv')
abbrevs = pd.read_csv('state-abbrevs.csv')
```

```python
display(pop.head(), areas.head(), abbrevs.head())
```

```python
# rank US states by their 2010 population density (density = population/ area)
```

```python
df3 = pd.merge(pop, abbrevs, left_on='state/region', right_on='abbreviation')
df3.head()
```

```python
df3 = df3.drop('abbreviation', axis=1)
df3.head()
```

```python
# check for null vals
df3.isnull().any()
```

```python
display(df3.head(), areas.head())
```

```python
df4 = pd.merge(df3, areas)
df4.head()
```

```python
# check for null vals
df4.isnull().any()
```

```python
data_2010 = df4.query("year == 2010 & ages == 'total'")
data_2010.head()
```

```python
# re-index data on state col
data_2010.set_index('state', inplace=True)
data_2010.head()
```

```python
final_df = data_2010['population']/ data_2010['area (sq. mi)']
final_df.head()
```

```python
final_df.sort_values(ascending=False, inplace=True)
final_df.head()
```
