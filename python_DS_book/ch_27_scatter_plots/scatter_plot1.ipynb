```python
%matplotlib inline
import matplotlib.pyplot as plt 
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'v', color='black');
```

```python
# multiple symbols
rng = np.random.default_rng(0)
for marker in ['o', '.', ',', 'x','+', 'v','^', '<', '>', 's', 'd']:
    plt.plot(rng.random(2), rng.random(2), marker, color='black', label="marker")
    
plt.legend(numpoints=1, fontsize=13) # numpoints - The number of marker points in the legend when creating a legend
plt.xlim(0, 1.8);
```

```python
# Scatter plots with plt.scatter
plt.scatter(x, y, marker='o');
```

```python
rng = np.random.default_rng(0)
x = y = rng.normal(size=100) # Draw random samples from a normal (Gaussian) distribution. Bell curve
colors = rng.random(100)
sizes = 1000 * rng.random(100)

plt.scatter(x, y,c=colors, s=sizes, alpha=.3) #The alpha blending value, between 0 (transparent) and 1 (opaque).
plt.colorbar();

```

```python
from sklearn.datasets import load_iris

iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=.4, s=100*features[3], c=iris.target, cmap='viridis')

plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```
