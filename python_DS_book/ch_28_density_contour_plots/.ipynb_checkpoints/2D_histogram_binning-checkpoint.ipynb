```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np 
plt.style.use('seaborn-white')
```

```python
rng = np.random.default_rng(1701)
```

```python
mean = [0, 0]
cov = [[1, 1], [1, 2]]

'''
Draw random samples from a multivariate normal distribution.
mean : 1-D array_like, of length N. Mean of the N-dimensional distribution
cov : 2-D array_like, of shape (N, N). Covariance matrix of the distribution
'''
x, y = rng.multivariate_normal(mean, cov, 10000).T
```

```python
rng.multivariate_normal?
```

```python
# 2D histogram : tessalation of squares
plt.hist2d(x, y, bins=30)
cb = plt.colorbar()
cb.set_label('counts in bin')
```

```python
counts, xedges, yedges = np.histogram2d(x, y, bins=30)
counts.shape
```

```python
#Hexagonal binning : tessalation of hexagons
plt.hexbin(x, y, gridsize=30)
cb = plt.colorbar(label='counts in bin')
```

```python
# kernel density estimation : estimate and represent densities in multiple dimensions : produces smoothing lenth
```

```python
from scipy.stats import gaussian_kde

#fit an array of size[Ndim, Nsamples]
data = np.vstack([x, y])
kde = gaussian_kde(data)

#evaluate on a regular grid
xgrid = np.linspace(-3.5, 3.5, 40)
ygrid = np.linspace(-6, 6, 40)
Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)
Z = kde.evaluate(np.vstack([Xgrid.ravel(), Ygrid.ravel()]))

#Plot the result as an image
plt.imshow(Z.reshape(Xgrid.shape), origin='lower', aspect='auto', extent=[-3.5, 3.5, -6, 6])
plt.colorbar(label='density');
```
