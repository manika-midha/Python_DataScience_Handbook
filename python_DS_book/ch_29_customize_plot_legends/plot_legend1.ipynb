```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np 
```

```python
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-b', label='Sine')
ax.plot(x, np.cos(x), '--r', label='Cosine')
ax.axis('equal')
ax.legend();
```

```python
ax.legend(loc='upper left')#frameon : Whether the legend should be drawn on a patch (frame).
fig
```

```python
ax.legend(loc='upper left', frameon=True)#frameon : Whether the legend should be drawn on a patch (frame).
fig
```

```python
#specify num of cols in legend
ax.legend(loc='lower center', ncol=2)
fig
```

```python
ax.legend(frameon=True,
          fancybox=True,
          framealpha=1,
          shadow=True,
          borderpad=1
         )
fig
```

```python
cities = pd.read_csv('california_cities.csv')
```

```python
#Extract data we are interested in
lat, lon = cities['latd'], cities['longd']
population, area = cities['population_total'], cities['area_total_km2']

# scatter the points using size and color but no label
'''
c = list of colors
s = marker size in points ** 2
linewidth = The linewidth of the marker edges
alpha = The alpha blending value, between 0 (transparent) and 1 (opaque).
'''
plt.scatter(lon, lat, label=None, c=np.log10(population), cmap='viridis', s=area, linewidth=0, alpha=.5)

plt.axis('equal')
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.colorbar(label='log$_{10}$(population)')
plt.clim(3, 7)

# create legend
# plot empty lists with the desired size and label
for area in [100, 300, 500]:
    plt.scatter([],[], c='k', alpha=.3, s=area, label=str(area) + ' km$^2$')
    
plt.legend(scatterpoints=1, frameon=False, labelspacing=1, title='City Area')

plt.title('California Cities: Area and Population');
```
