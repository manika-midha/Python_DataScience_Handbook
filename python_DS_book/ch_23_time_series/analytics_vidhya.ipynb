```python
from pandas_datareader import data as pdr
import yfinance as yf 
yf.pdr_override()
```

```python
#Magnificient 7 stocks
tickers = ['AAPL', 'AMZN', 'GOOGL', 'META', 'MSFT', 'TSLA', 'NVDA'] 
```

```python
df = pdr.DataReader(tickers, start='2020-01-01', end='2023-12-28', data_source='yahoo') #data_source='yahoo'
df.head() 
```

```python
start_date = '2020-01-01'
end_date = '2023-12-28'
source = 'yahoo'
```

```python
dfs = [] 
for sym in tickers:
    df = pdr.DataReader(sym, start=start_date, end=end_date, data_source='yahoo')
    dfs.append(df)
#     print(f"DF for {sym} is :")
#     print(df)
    
combined_df = pd.concat(dfs, axis=1)

```
