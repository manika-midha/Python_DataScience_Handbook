```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
data = pd.read_csv('FremontBridge.csv', index_col='Date', parse_dates=True)
data.head()
```

```python
data.columns = ['Total', 'East', 'West']
data.head()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.dropna().describe()
```

```python
data.plot()
plt.ylabel('Hourly Bicycle count')
```

```python
# resample by week
%matplotlib inline
import seaborn
seaborn.set()

#data.resample('W').sum().plot(style=['-', ':', '--'])
weekly = data.resample('W').sum()  # add vals for each week
weekly.plot(kind = 'line', style=['-', ':', '--'], figsize=(16, 6)) # line plot is default
plt.ylabel('Weekly bicycle count');
```

```python
weekly = data.resample('W').sum()
print('weekly size :',weekly.size)
weekly.plot(kind='line', figsize=(16,6))
plt.ylabel('Weekly Bicycle Count')
```

```python
# 30 day rolling mean to aggregate data (centre the window)
daily = data.resample('D').sum()
#daily.head()
daily.rolling(30, center=True).sum().plot(style=['-', ':', '--'])
plt.ylabel('mean hourly count.')
```

```python
# get smoother version of rolling mean using window func - Gaussian window
# width of window = 50 days, width of Gaussian window = 10 days

daily.rolling(50, center=True, win_type='gaussian').sum(std=10).plot(style=['-', ':', '--']);
```

```python
data.index.time
```

```python
# avg traffic as a function of the time of day
by_time = data.groupby(data.index.time).mean()
hourly_ticks = 4 * 60 * 60 * np.arange(6) # interval of 4 hrs in secs

print(by_time.head())
print(hourly_ticks)
```

```python
by_time.plot(xticks=hourly_ticks, style=['-', ':', '--']);
```
